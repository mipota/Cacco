### 仕事内容ごとの週の最低出勤日数の割合
- 仕事内容ごとにデータ数を集計(work_{i}day)
- day(0,1,2.5,4)の数を集計してwork_{i}dayで割って割合を出す
- 棒グラフ作成


```python
# groupbyメソッドで、'仕事内容'列ごとに'day'の数をカウントする
work_day = zikyu_df.groupby(['仕事内容'])['day'].count()
work_0day = work_day[2] 
work_1day = work_day[5]
work_2day = work_day[3]
work_3day = work_day[12]
work_4day = work_day[10]
work_5day = work_day[6]
work_6day = work_day[7]
work_7day = work_day[1]
work_8day = work_day[11]
work_9day = work_day[9]
work_10day = work_day[4]
work_11day = work_day[8]
work_12day = work_day[0]
work_0day
```

```python
# groupbyメソッドで、'仕事内容'ごとに'day'数を合計し、相対的な割合を算出
day_per = zikyu_df.groupby(['仕事内容'])['day'].value_counts(normalize=True)
day_per
```

```python
# groupbyメソッドで、'仕事内容'列ごとに'money'の数をカウントし
#最小値、最大値、平均値を算出する
money_work = zikyu_df.groupby(['仕事内容'])['money'].agg(['min','max','median','mean'])
newa = zikyu_df.groupby(['仕事内容'])['money'].quantile(0.25)
newb = zikyu_df.groupby(['仕事内容'])['money'].quantile(0.75)
df_s_h_in = pd.concat([money_work, newa,newb], axis=1, join='inner')

df_s_h_in.to_csv("money_work.csv", encoding="shift_jis")
```
